```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_file = "202209171645-ExportLogExam(3utf).csv"

sorted_difficulty = ["BASIC_CAL", "CONCEPT", "APPLICATION", "WORD_PROBLEM", "HELUSTIS"]
```

# Plotting S Curve of Each Level of Difficulty 

If it is in the shape of an S curve, it means the CAT model is working (it attempts to fit the students into each trait level, with the threshold divided at the center of the S).

```python
df = pd.read_csv(csv_file)
# Filter 1 Difficulty Level
df = df[df.levelOfDiffculty == sorted_difficulty[0]]
df
```

## Unique Student-Lesson-Skill-testRound

```python
colList = ['studentName', 'lessonId', 'skillId', 'testRound', 'traitLevelEstimate'] + ['className']
```

```python
df_unique = df.groupby(colList).size().reset_index(name='totalCount') \
            .sort_values(colList, ascending = [True] * len(colList) ) 
df_unique
```

## Find NumCorrect and Probability Correct

```python
df_unique_correct = df[df.isCorrect == True].groupby(colList).size().reset_index(name='totalCorrect')\
            .sort_values(colList, ascending = [True] * len(colList) ) 
df_unique_correct
```

```python
merged_df = pd.merge(df_unique, df_unique_correct, how='outer', on=colList)
# Make all NaN to 0
merged_df.loc[merged_df.totalCorrect.isnull(), 'totalCorrect'] = 0
merged_df
```

```python
# Calculate Correct Probability
merged_df['correctProb'] = merged_df.apply(lambda row: 0 if(row.totalCount == 0) else row.totalCorrect/ row.totalCount, axis=1)
merged_df
```

## S Curve Level Basic -2

```python
plt.figure(figsize=(13,7))
plt.scatter(merged_df.traitLevelEstimate.tolist() , merged_df.correctProb.tolist())
```
